# Sepsis와 non_sepsis 구분

```python
# sepsis와 non sepsis 구분

s = 0
e = 20
sepsis_data = 0
non_sepsis_data = 0


sepsis_df = pd.DataFrame(columns=cols)
non_sepsis_df = pd.DataFrame(columns=cols)

for _ in tqdm(range(people)):
    for i in range(17):
        if df_patient_per_20[s+i:s+i+3]['SepsisLabel'].sum() >= 1:
            #. sepsis data count
            sepsis_data += 1
            
            # new DF creat
            data = df_patient_per_20[s+i:s+i+3]
            sepsis_df = pd.concat([sepsis_df, data])
        else: 
            # non_sepsis data count
            non_sepsis_data += 1
            
            # new DF creat
            data = df_patient_per_20[s+i:s+i+3]
            non_sepsis_df = pd.concat([non_sepsis_df, data])
        
    s += 20
    e += 20
```

# sepsis vs non_sepsis 데이터 수 비교

```python
ratio_sepsis = round(sepsis_data/(sepsis_data+non_sepsis_data), 2)
ratio_non_sepsis = round(non_sepsis_data/(sepsis_data+non_sepsis_data), 2)


print(f'sepsis 발병 데이터 수 : {sepsis_data} ({ratio_sepsis})  || nonSepsis 데이터 수 : {non_sepsis_data} ({ratio_non_sepsis})')
```

# 환자별 feature별 평균값 추출

```python
cols = sepsis_df_no_label.columns
cols
```

```python
# 환자 별 평균
person_mean = []

s_i = 0
e_i = 20

for _ in tqdm(range(int(sepsis_df_no_label.shape[0]/20))):
    temp = []
    for col in cols:
        temp.append(sepsis_df_no_label[col][s_i:e_i].mean())
    person_mean.append(temp)
    
    s_i += 20
    e_i += 20
    
len(person_mean)
```

```python
# 환자별 평균을 정리한 리스트에 결측값을 0으로 변환

for i in tqdm(range(len(person_mean))):
    for j in range(len(person_mean[i])):
        if str(person_mean[i][j]) == "nan":
            person_mean[i][j] = 0
        else:
            person_mean[i][j] = round(person_mean[i][j], 1)
```

```python
# DP 생성 후 결측치 없는 데이터 세트 구성

sepsis_df_no_label_and_mv = pd.DataFrame(columns = cols)

s_i = 0
e_i = 20

for i in tqdm(range(sepsis_df_no_label.shape[0]//20)):
    temp = pd.DataFrame()
    for j, col in enumerate(cols):
        temp[col] = sepsis_df_no_label[col][s_i:e_i].fillna(person_mean[i][j])
        
    sepsis_df_no_label_and_mv = pd.concat([sepsis_df_no_label_and_mv, temp])
    
    s_i += 20
    e_i += 20
```

```python
def ratio(li):
    l = len(li)
    train = l*0.9
    return int(train), l
```

```python
tr, te = ratio(s_label)
print(tr, te)
```

```python
# sepsis 데이터 > x_train, x_test

s = 0
x_train = []
x_test = []

for _ in tqdm(range(int(sepsis_df_no_label_and_mv.shape[0]/20))):
    if _ < tr:
        for i in range(17):     # 20개의 데이터 세트에서 17개의 데이터를 label에 달아 사용 가능
            temp = sepsis_df_no_label_and_mv[s+i : s+i+3].values.tolist()
            x_train.append(temp)
    else: 
        for i in range(17):     # 20개의 데이터 세트에서 17개의 데이터를 label에 달아 사용 가능
            temp = sepsis_df_no_label_and_mv[s+i : s+i+3].values.tolist()
            x_test.append(temp)
    s += 20
```
